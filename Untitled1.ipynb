DATA CLEANING OF TRAINING SET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn
```

```python
df=pd.read_csv('training_set_features.csv',index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['xyz_concern']=df['xyz_concern'].fillna(2.0)
df['xyz_knowledge']=df['xyz_knowledge'].fillna(1.0)
df['behavioral_antiviral_meds'].value_counts()
df['behavioral_antiviral_meds']=df['behavioral_antiviral_meds'].fillna(0.0)
df['behavioral_avoidance']=df['behavioral_avoidance'].fillna(df['behavioral_avoidance'].mode()[0])
df['behavioral_face_mask']=df['behavioral_face_mask'].fillna(df['behavioral_face_mask'].mode()[0])
df['behavioral_wash_hands']=df['behavioral_wash_hands'].fillna(df['behavioral_wash_hands'].mode()[0])
df['behavioral_large_gatherings']=df['behavioral_large_gatherings'].fillna(df['behavioral_large_gatherings'].mode()[0])
df['behavioral_outside_home']= df['behavioral_outside_home'].fillna(df['behavioral_outside_home'].mode()[0])
df['behavioral_touch_face']= df['behavioral_touch_face'].fillna(df['behavioral_touch_face'].mode()[0])
df['doctor_recc_xyz'].value_counts()
#df.dropna(subset=['education','marital_status','rent_or_own','income_poverty','employment_industry','employment_occupation'],inplace=True)
df['employment_status']=df['employment_status'].fillna('Employed')
df['household_adults']=df['household_adults'].fillna(1.0)
df['household_children']=df['household_children'].fillna(0.0)
df['child_under_6_months']=df['child_under_6_months'].fillna(0.0)
df['health_worker']= df['health_worker'].fillna(df['health_worker'].mode()[0])
df['opinion_seas_sick_from_vacc']=df['opinion_seas_sick_from_vacc'].fillna(1.0)
df['opinion_xyz_sick_from_vacc']=df['opinion_xyz_sick_from_vacc'].fillna(2.0)
df['opinion_seas_risk']=df['opinion_seas_risk'].fillna(2.0)
df['opinion_xyz_risk']=df['opinion_xyz_risk'].fillna(2.0)
df['opinion_seas_vacc_effective']=df['opinion_seas_vacc_effective'].fillna(4.0)
df['opinion_xyz_vacc_effective']=df['opinion_xyz_vacc_effective'].fillna(4.0)
df['doctor_recc_seasonal']=df['doctor_recc_seasonal'].fillna(0.0)
df['doctor_recc_xyz']=df['doctor_recc_xyz'].fillna(0.0)
df['health_insurance']=-df['health_insurance'].fillna(1.0)
df['chronic_med_condition']=df['chronic_med_condition'].fillna(0.0)
df=df.drop(['employment_occupation','employment_industry','census_msa','hhs_geo_region','employment_status','rent_or_own','marital_status','race','education'],axis=1)
df=pd.get_dummies(df, columns=['sex', 'age_group','income_poverty'])
df.replace({True: 1.0, False:0.0}, inplace=True)
#df['respondent_id']=df['respondent_id'].astype(float)
```

```python
dx=dp.drop(['respondent_id','seasonal_vaccine'],axis=1)
```

```python
dx.head()
```

```python
X_tran=df
y_tran=dx
X_test=dt
```

```python
from sklearn.model_selection import train_test_split
#X_tran, X_test, y_tran, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_tran, y_tran)
y_xtest=model1.predict(X_test)
pred_prob2 = model1.predict_proba(X_test)
at=pd.DataFrame(y_xtest)
xyz=pd.DataFrame(pred_prob2)
xyz.head()
```

COMPUTING ROC AUC SCORE FOR BOTH VACCINES

```python
from sklearn.metrics import roc_curve


fpr1, tpr1, thresh1 = roc_curve(y_stest, pred_prob1[:,1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(y_xtest, pred_prob2[:,1], pos_label=1)
random_probs = [0 for i in range(len(y_xtest))]
p_fpr, p_tpr, _ = roc_curve(y_xtest, random_probs, pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score


auc_score1 = roc_auc_score(y_stest, pred_prob1[:,1])
auc_score2 = roc_auc_score(y_xtest, pred_prob2[:,1])

print(auc_score1, auc_score2)
```
